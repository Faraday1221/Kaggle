#### Beds & Baths & Rooms (done)
We can create a few new metrics for Beds and Baths that are more intuitive for home listings. i.e.
- Number of Bedrooms (total)
    - Estimated number of Basement BR
- Number of Bathrooms (total)
- Total Living SF (not counting porches)
- Total SF with Porches (no garage)
- Ratio of porch SF compared with Interior SF
- Percent of Finished SF in Basement (ratio)
- Percent of Living Space Above Ground (ratio)
- Rooms other than BR,BA and Kitchen (ratio)

```python
# total bedrooms

# About 1/3 of the houses have 1/2 to 1/3 of their finished living space in the basement
# x = pd.qcut(df.LvgSpaceAbvGr_Ratio,q=3)
# x.value_counts()

import math

# we need a metric for potential bedrooms in the basement
# we will guess that if there is >1 BA in the basement there is 1 BR, >2 BA = 2BR
df['Bsmt_BR'] = df['BsmtFullBath'].fillna(0)*1.0 + df['BsmtHalfBath'].fillna(0)*0.5
df['Bsmt_BR'] = df.Bsmt_BR.apply(lambda x: math.floor(x))

# display(df[['Bsmt_BR','BsmtFullBath','BsmtHalfBath']].head(20))
# display(df.Bsmt_BR.value_counts())

df['Total_BR'] = df['BedroomAbvGr'] + df['Bsmt_BR']

# df.Total_BR.value_counts()
```

```python
# total bathrooms
df['Total_BA'] = df['FullBath'] + df['HalfBath']*0.5 + df['BsmtFullBath'].fillna(0) + df['BsmtHalfBath'].fillna(0)*0.5


# Rooms other than bedrooms bathrooms and kitchen (as %)
df['Non_BA_BR'] = (df['BedroomAbvGr'] + df['FullBath'] + df['HalfBath'] + 1) / (df['TotRmsAbvGrd'])
```

```python
# ratio of basement finished sf to total sf
df['BsmtFinSF_Ratio'] = 1 - df['BsmtUnfSF'] / df['TotalBsmtSF']
df['BsmtFinSF_Ratio'].fillna(0,inplace=True)

# total living area
df['TotFinSF'] = df.TotalBsmtSF.fillna(0) - df.BsmtUnfSF.fillna(0) + df.GrLivArea

# ratio of living space that is above ground
df['LvgSpaceAbvGr_Ratio'] = 1 - (df['TotalBsmtSF'].fillna(0) - df['BsmtUnfSF'].fillna(0)) / df['TotFinSF'] 

# total SF considering porches etc.
df['All_SF'] = df['TotFinSF'] + df['ScreenPorch'] + df['3SsnPorch'] + df['EnclosedPorch'] \
                + df['OpenPorchSF'] + df['WoodDeckSF']

# df[['ScreenPorch','3SsnPorch','EnclosedPorch','OpenPorchSF','WoodDeckSF']].head(20)

# additional SF from porches etc.
df['PorchSF_Ratio'] =  (df.All_SF - df.TotFinSF) / df.TotFinSF

# df[['All_SF','TotFinSF','PorchSF_Ratio']].head(20)
```
